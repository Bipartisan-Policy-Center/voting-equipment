```python
df = pd.DataFrame([x.split('\t') for x in caulfield_appendix.strip().split('\n')[1:]], columns=caulfield_appendix.strip().split('\n')[0].split('\t'))

df['Median Price'] = df['Median Price'].replace('[$,]', '', regex=True).astype(float)
df['Number of Machines'] = df['Number of Machines'].astype(int)

# Group by Type and compute weighted average
weighted_avg_price = (
    df.groupby('Type')
    .apply(lambda x: (x['Median Price'] * x['Number of Machines']).sum() / x['Number of Machines'].sum(), include_groups=False)
    .reset_index(name='Weighted Average Median Price')
)

# Optional: format as currency
weighted_avg_price['Weighted Average Median Price'] = weighted_avg_price['Weighted Average Median Price'].map('${:,.0f}'.format)

weighted_avg_price
```
